Vamos analisar a situação de trabalho (vínculo empregatício) com a média salarial dos entrevistados.


```
raw_dados['mdsal'] = raw_dados["('P16', 'salary_range')"].fillna('$ 0/').apply(lambda x: #get the mean in ranges
                    int( #transform all in int in the end
                    (int(str(x)[str(x).rfind(' ')+1:str(x).rfind('/')].replace('.','')) # Get max in range 
                    +
                    int(str(x)[str(x).find('$')+2:str(x).find('/')].replace('.','')) #Get min
                    )/2)) #divide by 2
```

```
raw_dados.groupby("('P10', 'job_situation')")['mdsal'].head()

```

Contando com as exceções das categorias, que possuem salários maiores de 20 mil reais, temos que a média salárial de um servidor público é de R$8575, 00 e em segundo lugar, o empreendedor com a média de 8376 reais.

Vamos relacionar agora sobre a média salarial comparando o mesmo tempo de expêriencia entre homens e mulheres


```
raw_dados.groupby("('P2', 'gender')")['mdsal'].mean()
```

```
raw_dados.groupby("('P2', 'gender')")['mdsal'].mean().plot.bar(color=['#f5005a', 'blue'])
```

```
col_ad = "('P17', 'time_experience_data_science')"
raw_dados.loc[raw_dados[col_ad] == 'Não tenho experiência na área de dados', col_ad] = 'Sem experiência'

men = raw_dados[raw_dados.loc[:, "('P2', 'gender')"] == 'Masculino'] #Linhas com sexo Masculino
men_state_mean = men.groupby("('P5', 'living_state')")['mdsal'].mean() #Média Masculina por região

woman = raw_dados[raw_dados.loc[:, "('P2', 'gender')"] == 'Feminino'] #Linhas com sexo Feminino
woman_state_mean = woman.groupby("('P5', 'living_state')")['mdsal'].mean() #Média Feminina por região

menxp = men.groupby("('P17', 'time_experience_data_science')")['mdsal'].mean()
womanxp = woman.groupby("('P17', 'time_experience_data_science')")['mdsal'].mean()
```

```
plt.barh(men_state_mean.index, men_state_mean.values, color='blue')
plt.barh(woman_state_mean.index, woman_state_mean.values, color='#f5005a')

plt.xlabel("Média Salarial em R$")
plt.ylabel("Estados")

plt.title("Comparativo entre o salário de Homens e Mulheres por Região")

plt.legend(['Homens', 'Mulheres'])
```

```
plt.barh(menxp.index, menxp.values, color='blue')
plt.barh(womanxp.index, womanxp.values, color='#f5005a')

plt.xlabel("Média Salarial em R$")
plt.ylabel("Experiência na área de dados")
plt.legend(['Homens', 'Mulheres'])

plt.title("Comparativo entre o salário de Homens e Mulheres com mesma experiência")
```

De acordo com os gráficos acima, o primeiro dividido por regiões e o segundo por experiência, os homens ganham mais que as mulheres, mesmo com o mesmo tempo de experiência. 
Podemos observar que a diferença salarial entre os participantes de 4 a 5 anos e de mais de 10 anos são bem discrepantes. Porém, se analisarmos a experiência de menos de 1 ano e de 6 a 10 anos, essa diferença é mínima. Porém, existe uma diferença, o que preocupa bastante para a inserção de mais mulheres na área.
Considerando por estado, no Espírito santo a diferença salarial é mais de 50% e no RJ podemos ver que as mulheres possuem mesmo salário ou ganham mais que os homens.
